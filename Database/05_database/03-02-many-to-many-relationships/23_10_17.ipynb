### dumpdata -> 생성 데이터 추출

  - 에를들어 . python manage.py dumpdata --indent 4 articles.comment > comments. json 식
  - 주의할점은 에러를 방지하기 위해 생서할때
    python -Xutf8 manage.py dumpdata --indent 4 articles.comment > comments. json
  

### fixture 파일 기본경로
  - app_name/fixtures/
  - 파일 안보일시 폴더를 직접 찾아서 메모장으로 열고 UTF-8로 다른이름으로 저장 실행


### loaddata 활용
  - 모델 관계에서 user -> article -> comment 순으로 data를 넣어야 오류가 발생하지않음

  - python manage.py loaddata users.json(파일이름) 식으로 로드 가능

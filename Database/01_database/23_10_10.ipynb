### 데이터베이스 역할 : 데이터를 저장하고 조작 CRUD
  1. 관계형 데이터베이스 - 특정날짜에 구입한 고객, 아직 못받은 고객 등
    - 테이블, 행 열의 정보를 구조화하는 방식
    - 서로 관련된 데이터 포인터를 저장하고 이에 대한 액세스를 제공
    - 각 데이터에 고유한 식별값 부여(기본 키, Primary key) 고객정보
    - 주문데이터가 테이블에 저장되어 있다고 가정 고객의 고유한 식별값을 저장하자 즉(외래 키, Foreign Key) 고객 고유한 식별값


### 관계현 데이터베이스 키워드
  1. 테이블(Table) : 데이터를 기록하는곳
  2. 필드(Field) : 각 필드에는 고유한 데이터 형식이 지정됨
  3. 레코드(Record) : 각 레코드에는 구체저인 데이터 값이 저장됨
  4. 데이터베이스(Database) : 테이블의 집합
  5. 기본 키(Primary Key) : 각 레코드의 고유한 값으로 레코드의 식별자로 사용
  6. 외래 키(Foreign Key) : 다른 테이블의 레코드를 식별할 수 있는 키로 다른 테이블의 기본 키를 참조하고 각 레코드에서 서로 다른 테이블간의 관계를 만다는 데 사용

### SQL
  - 데이터베이스에 정보를 저장하고 처리하기 위한 프로그래밍 언어
  - 테이블의 형태로 구조화된 관계형 데이터베이스 에게 질의(요청) 하는것
  - SQL 키워드는 대소문자를 구분하지 않음 (하지만 대문자로 작성하는것을 권장)
  - 각 SQL Statements의 끝에는 세미콜론이 필요


### 수행 목적에 따른 SQL Statements 4가지 유형
 - DDL : 데이터의 구조 및 형식 변경
 - DQL : 데이터 검색
 - DML : 데이터 조작
 - DCL : 데이터 및 작업에 대한 사용자 권한 제어

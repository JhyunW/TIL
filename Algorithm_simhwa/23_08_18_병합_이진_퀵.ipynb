```python
# qick_ sort_lomuto
# 배열
# 왼쪽, 오른쪽(idx)
def quick_sort(arr, left, right):
    # 불할 정복의 가장 핵심
    if left < right:
        mid = cal(arr, left, right)
        quick_sort(arr, left, mid - 1)
        quick_sort(arr, mid + 1, right)

def cal(arr, left, right):
    # 피봇보다 큰 구간의 왼쪽 경계
    i = left - 1
    # 피봇보다 큰 구간의 오른쪽 경계
    j = left
    pivot = arr[right]
    while j < right:
        # 피봇이 j번째 값보다 더 크면
        if pivot > arr[j]:
            i = left - 1
            if i < j:
                # i와 j 사이 구간에 피봇보다 큰 값이 있다.
                arr[i], arr[j] = arr[j], arr[i]
        j += 1
    arr[i + 1], arr[right] = arr[right], arr[i+1]
    print(left,right)
    print(arr)
    return i + 1

nums = [11, 46, 23, 51, 28, 34]
print(nums)
quick_sort(nums, 0, len(nums) - 1)
```

```python
# 파이써닉 하게 만들어보기

def quick_sort(arr):
    # 분할
    if len(arr) <= 1:
        return arr
    else:
        # 분할 작업
        pivot = arr[0]
        left, right = [], []
        for i in range(1, len(arr)):
            if arr[i] > pivot:
                right.append(arr[i])
            else:
                left.append(arr[i])
        return [*quick_sort(left), pivot, *quick_sort(right)]

nums = [23, 12 60, 77, 32, 1]
```

```python
# 이진 탐색 재귀 활용
arr = [2, 4, 7, 9, 11, 23]

# 문제에서 데이터가 정렬되어 있다 라는 조건이 없다면
# 반드시 정렬을 먼저 수행해야 한다.

arr.sort()
low = 0
high = len(arr) - 1
mid = (low+high) // 2

# 함수 한 번 호출 때 마다 low, high 변수가 바뀌어서 사용됨
def binarySearch(target):
    # 기저 조건 : 언제까지 재귀호출을 반복할 것인가?
    if target == arr[mid]:
        return mid
    elif arr[mid] < target:
        return binarySearch(mid + 1, high, target)
    else:
        return binarySearch(low, mid - 1, target)
```

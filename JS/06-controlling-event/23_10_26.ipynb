## Event
  - 돔요소.addEventListener(수신할이벤트, 콜백함수) : 특정 이벤트를 DOM 요소가 수신할 때마다 콜백 함수를 호출
  즉 대상에 이벤트가 발생하면, 지정한 이벤트를 받아 할 일을 등록

## 버블링
  - 한 요소에 이벤트가 발생하면, 이 요소에 할당된 핸들러가 동작하고, 이어서 부모 요소의 핸들러가 동작
  - 가장 최산단의 요소를 만날때까지 반복, 요소 각각에 할당된 핸들러 동작
  - 이벤트가 제일 깊은 곳에 있는 요소에서 시작해 부모 요소를 거슬러 올라가며 발생하는 것이 마치 물속 거품과 닮았기 때문

## 타겟과 커렌트타켓
  - 타겟: 영향을 받은 위치들
    1. 이벤트가 발생한 가장 안쪽의 요소를 참조하는 속성
    2. 실제 이벤트가 시작된 타겟 요소
    3. 버블링이 진행되어도 변하지 않음
  
  - 커렌트타겟: 명확한 시작 위치
    1. 현재요소
    2. 항상 이벤트 핸들러가 연결된 요소만을 참조하는 속성
    3. this와 같음
